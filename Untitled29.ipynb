```python
# TASK 1: DATA LOADING & PREPROCESSING

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Generate Moons dataset
X, y = make_moons(n_samples=500, noise=0.25, random_state=42)

# Split data 70/30
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.30, random_state=42
)

# Standardize Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# Visualize original data
plt.figure()
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.title("Moons Dataset (Not Linearly Separable)")
plt.show()
```

```python
#TASK-2
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Linear SVM
linear_model = SVC(kernel='linear', C=1.0, random_state=42)
linear_model.fit(X_train_scaled, y_train)

# Predictions
linear_pred = linear_model.predict(X_val_scaled)

print("=== Linear SVM Report ===")
print(classification_report(y_val, linear_pred))
print("Accuracy:", accuracy_score(y_val, linear_pred))
```

```python
#TASK-3
# RBF SVM
rbf_model = SVC(kernel='rbf', random_state=42)
rbf_model.fit(X_train_scaled, y_train)
rbf_pred = rbf_model.predict(X_val_scaled)

print("=== RBF SVM Report ===")
print(classification_report(y_val, rbf_pred))

# Polynomial SVM
poly_model = SVC(kernel='poly', degree=3, random_state=42)
poly_model.fit(X_train_scaled, y_train)
poly_pred = poly_model.predict(X_val_scaled)

print("=== Polynomial SVM Report ===")
print(classification_report(y_val, poly_pred))
```

```python
print("Validation Accuracy Comparison:")
print("Linear SVM:", accuracy_score(y_val, linear_pred))
print("RBF SVM:", accuracy_score(y_val, rbf_pred))
print("Polynomial SVM:", accuracy_score(y_val, poly_pred))
```

```python
#TASK-4
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.1, 1, 10, 100],
    'kernel': ['rbf']
}

grid = GridSearchCV(
    SVC(random_state=42),
    param_grid,
    refit=True,
    verbose=2,
    cv=5,
    scoring='accuracy'
)

grid.fit(X_train_scaled, y_train)

print("=== Best Hyperparameters ===")
print(grid.best_params_)

print("=== Best CV Accuracy ===")
print(grid.best_score_)
```

```python
#TASK-5
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

final_predictions = grid.predict(X_val_scaled)

print("=== FINAL MODEL REPORT ===")
print(classification_report(y_val, final_predictions))

cm = confusion_matrix(y_val, final_predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix - Final Tuned Model")
plt.show()
```

```python
# Function to plot decision boundary
def plot_decision_boundary(model, X, y, title):
    x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
    y_min, y_max = X[:,1].min()-1, X[:,1].max()+1

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 200),
        np.linspace(y_min, y_max, 200)
    )

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
    plt.title(title)

# PLOT 1x3
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plot_decision_boundary(linear_model, X_train_scaled, y_train, "Linear SVM")

plt.subplot(1,3,2)
plot_decision_boundary(rbf_model, X_train_scaled, y_train, "RBF Kernel SVM")

plt.subplot(1,3,3)
plot_decision_boundary(grid.best_estimator_, X_train_scaled, y_train, "Tuned RBF (Best Model)")

plt.show()
```
